```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rate_small = pd.read_csv(r'C:\Users\Kevin\Desktop\bigdataproject\ml-latest-small\ratings.csv')
rate_large = pd.read_csv(r'C:\Users\Kevin\Desktop\bigdataproject\ml-latest\ml-latest\ratings.csv')
```

# Imbalanced data (small dataset)

```python
rate_small.groupby(rate_small['userId']).count()['movieId']
```

```python
num_data_s1 = list(rate_small.groupby(rate_small['userId']).count()['movieId'])
user_list_s1 = list(rate_small['userId'].unique())
plt.plot(user_list_s1, num_data_s1)
plt.xlabel('userId')
plt.ylabel('number of data')
plt.title('Imbalanced user data graph(small)')
```

```python
rate_small.groupby(rate_small['movieId']).count()['userId']
```

```python
num_data_s2 = list(rate_small.groupby(rate_small['movieId']).count()['userId'])
movie_list_s2 = list(rate_small['movieId'].unique())
plt.scatter(movie_list_s2, num_data_s2)
plt.xlabel('movieId')
plt.ylabel('number of data')
plt.title('Imbalanced movie data graph(small)')
```

```python
rate_small.groupby(rate_small['userId']).mean()['rating']
```

```python
avg_rate = list(rate_small.groupby(rate_small['userId']).mean()['rating'])
user_list_s3 = list(rate_small['userId'].unique())
plt.scatter(user_list_s3, avg_rate)
plt.xlabel('userId')
plt.ylabel('average of ratings for each user')
plt.title('Rating data(small)')
```

```python
sns.jointplot(x='userId',y='rating', data=rate_small,kind='hex')
```
